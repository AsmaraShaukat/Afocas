```python
# Install bertopic if you haven't already
# pip install bertopic

!pip install scikit-learn

from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

import zipfile
import os

# Path to the uploaded zip file
zip_file_path = "/content/archive (1).zip"  # Adjust the path accordingly

# Directory to extract the zip file contents
extracted_dir_path = "/content/sample_data/20_newsgroups"  # Adjust the path accordingly

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir_path)

# List the directories within the extracted directory
subdirectories = [os.path.join(extracted_dir_path, d) for d in os.listdir(extracted_dir_path) if os.path.isdir(os.path.join(extracted_dir_path, d))]

# Initialize a list to store the text data
text_data = []

# Load the text files from each subdirectory
for directory in subdirectories:
    files = [os.path.join(directory, f) for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    for file_path in files:
        with open(file_path, 'r', encoding='latin1') as file:
            text_data.append(file.read())

# Convert all elements in text_data to strings
text_data = [str(text) for text in text_data]

# Ensure that all elements in text_data are strings
if not all(isinstance(text, str) for text in text_data):
    raise TypeError("Make sure that the iterable only contains strings.")

# Initialize an empty list to store the text data
text_data = []

# Loop through the files in the extracted directory
for root, dirs, files in os.walk(extracted_dir_path):
    for file in files:
        # Check if the file has a .txt extension
        if file.endswith(".txt"):
            file_path = os.path.join(root, file)
            # Read the text file and append its content to the text_data list
            with open(file_path, 'r', encoding='latin1') as f:
                text_data.append(f.read())

# Ensure that all elements in text_data are strings
text_data = [str(text) for text in text_data]

# Initialize BERTopic model
model = BERTopic()

# Fit BERTopic model to the data
topics, _ = model.fit_transform(text_data)

# Get the most frequent topics
most_freq_topics = model.get_topic_freq()

# Visualize the topics
if isinstance(topics, list):
    # Convert topics to integers if they are not already
    topics = [int(topic) for topic in topics]

model.visualize_topics(topics=topics)

# Get the top n topics
top_topics = model.get_topics()

# Get the topic representation of a new document
new_document = "Your new document text goes here"
topic_id, _ = model.transform([new_document])

print("Topic ID of the new document:", topic_id[0])
print("Most frequent topics:", most_freq_topics)
print("Top topics:", top_topics)
```

```python

# Get the top topics with their associated words and weights
top_topics = model.get_topics()

# Define a function to extract topic names based on top words
def extract_topic_names(topics, num_words=5):
    topic_names = {}
    for topic_id, words in topics.items():
        # Join the top words into a single string
        topic_name = ", ".join(word for word, _ in words[:num_words])
        topic_names[topic_id] = topic_name
    return topic_names

# Extract topic names
topic_names = extract_topic_names(top_topics)

# Print the topic names
for topic_id, topic_name in topic_names.items():
    print(f"Topic {topic_id}: {topic_name}")
```
