**Preprocessing**

Converting string to list

```python
#Convert String to List
import ast

def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies["genres"] = movies['genres'].apply(convert)
movies["keywords"] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

**appending cast columns to first 3 names**

```python
def Convert(obj):
    List = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            List.append(i['name'])
            counter += 1
        else:
            break
    return List
```

```python
movies['cast'] = movies['cast'].apply(Convert)
movies.head(2)
```

**Extrating Director name from crew column**

```python
#From the crew,we are focusing mainly on directors for now
def fetch_director(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i ['job'] == "Director":
            List.append(i['name'])
    return List
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(2)
```

```python
#Convert other strings into list. Remove spaces
movies['overview'] = movies['overview'].apply(lambda x:x.split())

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(2)
```

 **Combining columns into one**

```python
#Combine all the columns
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

**Making New dataset with relevant columns**

```python
#Select the relevant columns
df = movies[["movie_id","title","tags"]]
df.head()
```

Importing poster stemmer from natural language toolkit which will help to stem words from text for us means it will convert or reduce words to their base form.

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
df['tags'] = df["tags"].apply(lambda x:" ".join(x))
```

```python
df['tags'][0]
```

Converting text to martix of token counts

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
vector
vector.shape
```


**Compute the cosine similarity between vectors**

```python
#Generate pairwise similarity matrix using cosing similarity
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
similarity.shape
```

**The Python function, named recommend, take a movie title as input and  recommend similar movies based on the computed similarity matrix and a DataFrame df**

```python
#Recommend top-5 movies

def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
